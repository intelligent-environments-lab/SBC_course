## kNN based classification

example is from https://towardsdatascience.com/implementing-k-nearest-neighbors-with-scikit-learn-9e4858e231ea

** 1. Read data **

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
```

```python
X
```

```python
type(X)
```

```python
y = iris.target
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=ListedColormap(['#FF0000', '#00FF00', '#0000FF']))
```

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

for i in range(20):
    n_neighbors = i+1
    
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i)"
              % n_neighbors)

plt.show()
```

## k-nn classification for building data

the data is from UC Irvine Machine Learning repository

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

%matplotlib inline
```

```python
eD = pd.read_csv('energyData.csv') 
```

```python
eD.head()
```

```python
plt.boxplot(eD['Y1'])
```

** 2. Separate the data **

```python
#google: np where two conditions

Y2 = np.where(eD['Y1'] > 30, 'high', (np.where(eD['Y1'] < 10, 'low', 'medium')))
```

```python
Y2
```

```python
set(Y2)
```

```python
X = eD.iloc[:, 0:2]
```

```python
X
```

```python
X.X1
```

```python
X.X2
```

```python
Y3 = np.where(eD['Y1'] > 30, 2, (np.where(eD['Y1'] < 10, 0, 1)))
```

```python
Y3
```

```python
plt.scatter(X.X1,X.X2,c=Y3,cmap=ListedColormap(['red', 'green', 'blue']))
```

** 3. Split the data & 4. Learn the model **

```python
from sklearn.model_selection import train_test_split  

X_train, X_test, y_train, y_test = train_test_split(X, Y2, test_size = 0.20)
```

```python
from sklearn import neighbors
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_validate

scores_2 = cross_validate(model, X_train, y_train, scoring='precision_macro', cv=5, return_train_score=False)
```

```python
scores_2
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

** 5. Evaluation **

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
cm = confusion_matrix(y_test, predictions)
print(cm)
```

```python
print(classification_report(y_test,predictions))
```
